```python
df = pd.read_csv("../data/nykaa_dataset_clean.csv")
```

```python
df.info()
```

```python
df["in_stock"] = (
    df["in_stock"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": True, "no": False})
)
```

```python
df["in_stock"] = df["in_stock"].fillna(False)
```

```python
df.info()
```

```python
df["in_stock"].value_counts(dropna=False)
```

```python
df["in_stock"] = (
    df["in_stock"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        "yes": True,
        "no": False,
        "true": True,
        "false": False,
        "in stock": True,
        "out of stock": False,
        "1": True,
        "0": False,
        "nan": False
    })
)
```

```python
df["in_stock"] = df["in_stock"].astype(bool)
```

```python
df.info()
```

```python
df.to_csv("../data/nykaa_cleaned.csv", index=False)
```

```python
df = pd.read_csv("../data/nykaa_cleaned.csv")

# Optional: create extra columns for analysis
df["discount"] = ((df["mrp"] - df["price"]) / df["mrp"] * 100).round(2)
df["price_bucket"] = pd.cut(df["price"], bins=[0, 200, 500, 1000, 1500, 2000, 5000], labels=["0-200","201-500","501-1000","1001-1500","1501-2000","2001-5000"])
```
